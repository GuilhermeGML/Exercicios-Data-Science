```python
agua = pokemon['Type 1']== 'Water'
agua1 = pokemon[agua]
agua1['Attack'].sum()
```

```python
tipos = ['Grass', 'Fire', 'Water', 'Bug', 'Normal', 'Poison', 'Electric', 'Ground', 'Fairy',
         'Fighting', 'Psychic', 'Rock', 'Ghost', 'Ice', 'Dragon', 'Dark', 'Steel', 'Flying']

for tipo in tipos:
    tipo_filter = pokemon['Type 1'] == tipo
    tipo_pokemon = pokemon[tipo_filter]
    total_attack = tipo_pokemon['Attack'].sum()
    print(f'Total de ataque para o tipo {tipo}: {total_attack}')
```

```python
plt.figure(figsize=(15, 5))
plt.title('Elementos dos Pokemons', fontsize=12)
sns.barplot(data=pokemon, x='Type 1', y='Attack', ci=None, estimator=sum)
```

```python
plt.figure(figsize=(15, 5))
plt.title('Elementos dos Pokemons', fontsize=12)
sns.barplot(data=pokemon, x='Type 1', y='Defense', ci=None, estimator=sum)
```

```python
plt.figure(figsize=(15, 5))
plt.title('Ataque dos Pokemons', fontsize=12)
sns.barplot(data=pokemon, x='Type 1', y='Attack', ci=None, estimator=sum)
plt.figure(figsize=(15, 5))
plt.title('Defesa dos Pokemons', fontsize=12)
sns.barplot(data=pokemon, x='Type 1', y='Defense', ci=None, estimator=sum)
```
